```python
import pandas as pd
```

```python
# load a_lvr_land_a.csv 轉成　dataframe：df_a
df_a = pd.read_csv('./lvr_land/a_lvr_land_a.csv', encoding='utf-8')
# 刪除所有的英文欄位
df_a.drop(index=0, axis=0, inplace=True)
df_a.reset_index(drop=True)
```

```python
# load b_lvr_land_a.csv 轉成　dataframe：df_b
df_b = pd.read_csv('./lvr_land/b_lvr_land_a.csv', encoding='utf-8')
# 刪除所有的英文欄位
df_b.drop(index=0, axis=0, inplace=True)
df_b.reset_index(drop=True)
```

```python
# load e_lvr_land_a.csv 轉成　dataframe：df_e
df_e = pd.read_csv('./lvr_land/e_lvr_land_a.csv', encoding='utf-8')
# 刪除所有的英文欄位
df_e.drop(index=0, axis=0, inplace=True)
df_e.reset_index(drop=True)
```

```python
# load f_lvr_land_a.csv 轉成　dataframe：df_f
df_f = pd.read_csv('./lvr_land/f_lvr_land_a.csv', encoding='utf-8')
# 刪除所有的英文欄位
df_f.drop(index=0, axis=0, inplace=True)
df_f.reset_index(drop=True)
```

```python
# load h_lvr_land_a.csv 轉成　dataframe：df_h
df_h = pd.read_csv('./lvr_land/h_lvr_land_a.csv', encoding='utf-8')
# 刪除所有的英文欄位
df_h.drop(index=0, axis=0, inplace=True)
df_h.reset_index(drop=True)
```

```python
# 將 df_a, df_b, df_e, df_f, df_h 五張dataframe合併成df_all
df_all = pd.concat([df_a, df_b, df_e, df_f, df_h], axis=0, ignore_index=True)
```

```python
# 使用re過濾關鍵字
import re
def is_match_keywords(input_string, re_rules):
    re_match_name = re.compile(re_rules)
    if re_match_name.findall(str(input_string)) != []:
        # 有關鍵字符合
        return True
    else:
        return False
```

```python
# 比對關鍵字 住宅大樓
building_type = '住宅大樓'
# 比對關鍵字 十三~十九層、二十層~九十九層
floor_type = '十+[三四五六七八九]+層|[二三四五六七八九]+十+[一二三四五六七八九]+層|[二三四五六七八九]+十層'

#【主要用途 】為【住家用】 &【建物型態】為【住宅大樓】 &【總樓層數】需【大於等十三層】
filter_a = df_all[(df_all["主要用途"] == "住家用") & [is_match_keywords(s, building_type) for s in df_all['建物型態']] & [is_match_keywords(s, floor_type) for s in df_all['總樓層數']]]
# 將比對結果的dataframe存成filter_a.csv
filter_a.to_csv('./filter_a.csv', index = 0, encoding = 'utf_8_sig')
```

```python
df_all_2 = df_all.copy()

# 總件數
total_number = len(df_all_2)
total_number

# 總車位數
# 檢查交易筆棟數是否有任一筆為空值
df_all_2['交易筆棟數'].isnull().any() # -> False 無空值
parking_space_sum = sum([int(re.search("車位(\\d+)",n).group(1)) for n in df_all["交易筆棟數"]])

# 平均總價元
# 檢查總價元是否有任一筆為空值
# 計算平均總價元
df_all_2['總價元'].isnull().any() # -> False 無空值
total_price_avg = df_all_2['總價元'].astype(int).sum() / total_number

# 平均車位總價元
# 檢查車位總價元是否有任一筆為空值
df_all_2['車位總價元'].isnull().any() # -> False 無空值
# 平均車位總價元
total_parking_space_avg = df_all_2['車位總價元'].astype(int).sum() / parking_space_sum
```

```python
dict1 = {
    '總件數':total_number,
    '總車位數':parking_space_sum,
    '平均總價元':total_price_avg,
    '平均車位總價元':total_parking_space_avg 
}
dict1
```

```python
# 將結果組成的dict轉成dataframe
filter_b = pd.DataFrame(dict1, index = [0])
```

```python
# 將計算結果的dataframe存成filter_b.csv
filter_b.to_csv('./filter_b.csv', index = False, encoding = 'utf_8_sig')
```
